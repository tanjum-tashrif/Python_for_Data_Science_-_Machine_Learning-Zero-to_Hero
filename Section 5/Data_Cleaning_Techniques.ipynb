```python
import pandas  as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.Series([2.1,2.3,4.5,2.2,2.4])  #Median
median = np.median(x)
threshold = 2
outliers = []
for item in x:
  if abs(item - median) > threshold:  ####Rules
    outliers.append(item)
print(outliers)
```

```python
mean=np.mean(x)  #Mean
std=np.std(x)
outliers =[]
for item in x:
  if(item<mean-std) or (item>mean+std):   ####Rules
    outliers.append(item)
outliers
```

```python
mean=np.mean(x)  #Z_Score
std=np.std(x)
outliers =[]
for item in x:
  z_scores =(item-mean)/std
  if z_scores>1.5:
    outliers.append(item)
outliers
```

```python
###IQR
Q1,Q3 =np.percentile(x,[25,75])
IQR=Q3-Q1
outliers=[]
for item in x:
  if item<(Q1-1.5*IQR) or item>(Q3+1.5*IQR):
    outliers.append(item)
outliers
```
